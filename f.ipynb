```python
%matplotlib inline
import pylab as plt
import numpy as np
from numpy import tan, pi
```

```python
def linspace(n, plus=0):
    return np.linspace(-pi/2+n*pi+plus+0.001, pi/2+n*pi+plus-0.001, 1000)
def plotTan(n, label='', c='b'):
    x = linspace(n)
    y = tan(x)

    plt.plot(x, y, color=c, label=label)

def plotCot(n, label='', c='r'):
    x = linspace(n, plus=pi/2)
    y = -1/tan(x)

    plt.plot(x, y, color=c, label=label)

def plotVerticalLine(x0, c):
    if type(x0) != list and type(x0) != np.array:
        x0 = [x0]
        
    for x in x0:
        plt.plot([x, x], [-1000, 1000], '--', c=c)
        
        
def plotQK(vv, label='', c='g'):
    x = np.linspace(0.01, np.sqrt(vv)-0.0001, 1000)
    y = np.sqrt(vv/x**2-1)
    
    plt.plot(x, y, color=c, label=label)
```

```python
fig = plt.figure(figsize=(10,7))

plotTan(-1, label='$tan(ka)$')
plotTan(0)
plotTan(1)
plotCot(0, label='$cot(ka)$')
plotCot(1)
plotCot(-1)

plotQK(1)
plotQK(7)
plotQK(17)
# plotQK(20)
# plotQK(35, label=r'$q/k=\sqrt{\frac{V_0}{E}-1}=\sqrt{\frac{2mV_0}{\hbar^2k^2}-1}=\sqrt{\frac{2ma^2}{\hbar^2}\,\frac{V_0}{(ka)^2}-1}$')
plotQK(35, label=r'$q/k$')

plotVerticalLine([-3*pi/2, -pi/2, pi/2, 3*pi/2], c='b')
plotVerticalLine([-pi, 0, pi, 2*pi], c='r')
# plt.plot([-3*pi/2-0.1, 4*pi/2+0.1], [0, 0], '-', c='k')
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.ylim(-10,10)
plt.xlim(-3*pi/2-0.1, 4*pi/2+0.1)
# plt.legend(loc='best', fontsize=16)
# plt.xlabel('$ka$', fontsize=20)

ax = plt.gca()
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
# ax.set_position([box.x0, box.y0, box.width, box.height * 0.8])

plt.xticks([-3*pi/2, -pi/2, pi/2, 3*pi/2], ['$-3\pi/2$', '$-\pi/2$', '$\pi/2$', '$3\pi/2$'], fontsize=16)
plt.yticks([])
# for label in ax.get_xticklabels() + ax.get_yticklabels():
#     label.set_fontsize(16)
#     label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))
for tick in ax.xaxis.get_majorticklabels():
    tick.set_horizontalalignment("left")

ax.annotate('$ka$', xy=(0.95, 0.48), ha='left', va='top', xycoords='axes fraction', fontsize=20)
ax.annotate(r'$\frac{q}{k}=\sqrt{\frac{V_0}{E}-1}$', xy=(1.05, 0.7), ha='left', va='top', xycoords='axes fraction', fontsize=22)
ax.annotate(r'$=\sqrt{\frac{2mV_0}{\hbar^2k^2}-1}$', xy=(1.07, 0.55), ha='left', va='top', xycoords='axes fraction', fontsize=22)
ax.annotate(r'$=\sqrt{\frac{2ma^2}{\hbar^2}\,\frac{V_0}{(ka)^2}-1}$', xy=(1.07, 0.4), ha='left', va='top', xycoords='axes fraction', fontsize=22)
# ax.annotate('$x$', xy=(0, 1), xytext=(-15,2), ha='left', va='top', xycoords='axes fraction', textcoords='offset points', fontsize=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.35, 1.0), fontsize=18)

# ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.35), fontsize=18)

fig.tight_layout()
fig.savefig('fig_2-1.png', bbox_inches='tight', pad_inches=0.1, dpi=100)
```
